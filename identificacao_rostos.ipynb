```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread("/content/rostos-confianca.jpg")
imagemPB = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
```

```python
face_frontal = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
face_recon = face_frontal.detectMultiScale(imagemPB, scaleFactor=1.2, minNeighbors=1)

print('Faces Encontradas: ', len(face_recon))
```

```python
for (x, y, w, h) in face_recon:
  cv2.rectangle(imagem, (x,y), (x+w, y+h), (0, 255, 0), 2)

plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
```
